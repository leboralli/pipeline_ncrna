# Criando uma tabela com IDs, IDs completos, fenótipos, número de reads e etc

```python
import pandas as pd
import re
import glob
import os
```

```python
with open('/home/boralli/workdir/samples_30kkreads.csv') as f:
    fileSamples = pd.read_csv(f, sep=',')
fileSamples
```

```python
path = '/home/boralli/workdir/pipeline_v4/LOGS/FASTP_LOG/'
lista_nomes = []
pattern = r"R\d{4}_\w{9}_\w{6,8}_\w{4}_"
for filename in glob.glob(os.path.join(path, '*.html')):
#     lista_nomes.append(filename)
    regexList = re.findall(pattern, filename)
#     print (regexList)
    lista_nomes.extend(regexList)
        
lista_nomes = sorted(lista_nomes)

#adicionando a string _quant
string = '_quant'
lista_nomes_quant = [x + string for x in lista_nomes]

```

```python
path_salmon = '/home/boralli/workdir/pipeline_v4/SALMON/output/'
lista_salmon = []
pattern = r"R\d{4}_\w{9}_\w{6,8}_\w{4}__quant"
for filename in glob.glob(os.path.join(path_salmon)):
#     lista_nomes.append(filename)
    regexList = re.findall(pattern, filename)
#     print (regexList)
    lista_salmon.extend(regexList)
        
lista_salmon = sorted(lista_nomes)
```

```python
fileSamples['completeID'] = lista_nomes_quant

fileSamples
```
